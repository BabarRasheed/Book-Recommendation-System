## Popularity Base Recommendor System

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df
```

```python
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
import pandas as pd  # Import pandas for data manipulation

# Assuming 'ratings_with_name' is your DataFrame containing book ratings

# Check if 'Book-Rating' is numeric (optional but recommended)
if not pd.api.types.is_numeric_dtype(ratings_with_name['Book-Rating']):
    print("Warning: 'Book-Rating' column may contain non-numeric data.")

# Calculate average rating for each book title (assuming 'Book-Rating' is numeric)
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()

# Rename column if necessary (assuming successful mean calculation)
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)

avg_rating_df  # Print the resulting DataFrame with average ratings
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

## Collaborative Filtering Based Recommender System

```python
ratings_with_name
```

```python
ratings_with_name.groupby('User-ID').count()['Book-Rating']
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
x
```

```python
padhe_likhe_users = x[x].index
padhe_likhe_users
```

```python
ratings_with_name['User-ID'].isin(padhe_likhe_users)
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
filtered_rating
```

```python
filtered_rating.groupby('Book-Title').count()['Book-Rating']
```

```python
filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
y[y]    # Display book titles with 50 or more ratings from 'padhe_likhe_users'
```

```python
famous_books = y[y].index
famous_books
```

```python
filtered_rating['Book-Title'].isin(famous_books)
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(pt)
```

```python
similarity_scores =  cosine_similarity(pt)
```

```python
similarity_scores[0]
```

```python
similarity_scores.shape
```

```python
def recommend_books(book_name):
      index = np.where(pt.index == book_name )[0][0]
      similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[:10]

      data = []

      for i in similar_items:
            item = []
      
            temp_df = books[books['Book-Title'] == pt.index[i[0]]]
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
            data.append(item)

      return data

```

```python
recommend_books('1984')
```
